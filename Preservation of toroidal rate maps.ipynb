```python
def rot_para(params1,params2):   
    if np.abs(np.cos(params1[0])) < np.abs(np.cos(params2[0])):
        y = params1.copy()
        x = params2.copy()
    else:
        x = params1.copy()
        y = params2.copy()

    alpha = (y[0]-x[0])
    if (alpha < 0) & (np.abs(alpha) > np.pi/2):
        x[0] += np.pi 
        x[0] = x[0]%(2*np.pi)
    elif (alpha < 0) & (np.abs(alpha) < np.pi/2):
        x[0] += np.pi*4/3  
        x[0] = x[0]%(2*np.pi)
    elif np.abs(alpha) > np.pi/2:
        y[0] -= np.pi/3
        y[0] = y[0]%(2*np.pi)
    if y[0]>np.pi/2:
        y[0] -= np.pi/6
        x[0] -= np.pi/6
        x[0] = x[0]%(2*np.pi)
        y[0] = y[0]%(2*np.pi)
    if x[0]>np.pi/2:
        y[0] += np.pi/6
        x[0] += np.pi/6
        x[0] = x[0]%(2*np.pi)
        y[0] = y[0]%(2*np.pi)
    return x,y
```
